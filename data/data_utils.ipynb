```python
# Converts the LIMA JSONL file to a CSV file of questions only
#   and samples n_samples if provided
# Download from https://huggingface.co/datasets/GAIR/lima

import pandas as pd
from pathlib import Path

def lima_jsonl_to_questions_csv(jsonl_path: str, csv_path: str, n_samples: int = None, max_length: int = None):
    df = pd.read_json(Path(jsonl_path), lines=True)
    if max_length is not None:
        df = df[df["conversations"].apply(lambda x: len(x[0]) <= max_length)]
    questions = df["conversations"].apply(lambda x: x[0])
    if n_samples is not None:
        questions = questions.sample(n=n_samples, random_state=8128)
    questions.to_csv(csv_path, index=False, header=["prompt"])

lima_jsonl_to_questions_csv("lima_train.jsonl", "lima_train_subset.csv", n_samples=50, max_length=800)
```
